### Sequences:
They can be lists, tuples or strings. They are objects that are ordered and they can be mutable or inmutable.

In python we can access a string as we would a list. 

```python
sentence = "This is a sequence"
sentence[10:18]
```

This could come really handy for some applications, like:

```python
sentence = "This is a sequence"
print(sentence[::-1])
```

We can also slice an entire sequence within a sequence and replace with another sequence. We just need to asign the sequence on the reight hand sign to the sequence slice of the list that we want to replace and it doesnt matter if the sequence we want to replace with is a tuple a list or a string. 

```python
a_list =  [1, 2, 3, 4, 5]
a_list[0:2] = ['a', 'b']
print(a_list)
a_list[0:3] = ('c', 'd')
print(a_list)
```

### Step sequence replacement
we can use slicing to do some useful operations like replacement with a step, or we could also go backwards, but if we offer a step we need to set a sequence to replace the slice the same size. for example:

```python
l = [1, 2, 3, 4, 5, 6, 7, 8]
l[1::2] = 20, 40, 60, 80
l
```

we could also step and replace like this

```python
l = [1, 2, 3, 4, 5, 6, 7, 8]
l[::2]
```
