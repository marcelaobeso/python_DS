### Sequences:
They can be lists, tuples or strings. They are objects that are ordered and they can be mutable or inmutable.

In python we can access a string as we would a list. 

```python
sentence = "This is a sequence"
sentence[10:18]
```

This could come really handy for some applications, like:

```python
sentence = "This is a sequence"
print(sentence[::-1])
```
