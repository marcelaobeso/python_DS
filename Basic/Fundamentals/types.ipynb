### Comparison Operators

In python we have ==, !=, <=, >= to compare value equality 

But we also have identity comparison and here we will use 'is'


````python
a = 1
b = 1.0

print(a == b)
print(a is b)
````

but we could have problems with certain values, because of some python interpreter optimizations, for small numbers python could reuse the precreated object that already has an space set in memory, for example 

````python
a = 1
b = 1
print(a is b)
id(a), id(b)
````

We also have membership comparison operators, that checks if a value is part of a collection or a set of values, for example: 

````python
values = { 1, 3, 4, 'cat'}
print('cat' in values)
````

### Booleans

to compare thruthness we have _and_, _or_ and _not_ we can use them as usual but python has another optimization called short-cirtuted evaluation where if we have an expresion such as:

```sin(a) > 0 and cos(a) < 0 ```

and we get that ```sin(a)``` is false, then inmediately instead of calculating it the interpreter short circuts and returns False to save the cost of evaluating the second part of the comparison and the 'and' 

same happens for "or" in the case the first operand is True.

For example this would throw an exception if it got to evaluate the second comparator but python shortcircuted on the first operand becuase of the and:
