With python we have several options like if.. else statements, elif to get more options to evaluate and then we have the Ternary operator. We could nest them but we should prioritize redability and the same with elif statements, maybe we could use a different structure. 

```python
a = 0 
b = 2
c = 4

if a < b: 
    print('A is not bigger than B')
    if b < c:
        print('b is bigger but not bigger than C')
    else:
        print('c is thebiggest number')
else: 
    print('A is bigger')
```

```python
grade = 72

if grade >= 90:
    letter_grade = 'A'
elif grade >= 80:
    letter_grade = 'B'
elif grade >= 70:
    letter_grade = 'C'
elif grade >= 60:
    letter_grade = 'D'
else:
    letter_grade = 'E'

print(letter_grade)
```
