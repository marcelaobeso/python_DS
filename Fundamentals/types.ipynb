### Floats
They are representations, the real value would not be an exact value, the kernel prints a string representation 

```python
0.1
```

for example, taking 0.1 we will get: 

```python
format(0.1, '.25f')
```

If we where to compare float numbers we could un into discrepancies, for example:

```python
0.1 + 0.1 + 0.1 == 0.3
```

```python
0.125 + 0.125 + 0.125 == 0.375
```

Python interpreter has some representation of certain decimals to handdle some definite fractions, but this comparisons might not work for all floats.
